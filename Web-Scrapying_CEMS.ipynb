```python
import pandas as pd
import requests
import json

# Set the URL for the JSON data
url = 'https://data.epa.gov.tw/api/v2/aqx_p_493?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=m_time%20desc&format=JSON'
response = requests.get(url)

# save as string type from request
d = response.text

# Parsing a string that resembles the JSON format into a dictionary or list.
data = json.loads(d)

print(data)
```

```python
# Get each item of records one by one
for i in data['records']:
    
    # print city and company name
    print(i['epb'] + ' ' + i['abbr'], end='，')
    
    # print the record time
    print('監測時間:' + i['m_time'], end='，')
    
    # print the record value
    print('監測數值:' + i['m_val'])
```

```python
# Use panda to create dataframe
df = pd.DataFrame(data['records'], columns=['epb', 'abbr', 'm_time', 'm_val'])
df.head()
```

```python
df.dtypes
```

```python
# change the 'm_time' from object into DateTime
df['m_time'] = pd.to_datetime(df['m_time'])

# change the object into float inorder to analyze
df['m_val'] = df['m_val'].replace('', 0).astype(float)

print(df.dtypes)
```
